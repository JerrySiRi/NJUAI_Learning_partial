```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal

HK = pd.read_csv("./h329b.csv",index_col=0)
HK_valid = HK[HK["Month"]==12].loc["2018"]
HK_valid["2018_12_Time"] = HK_valid["Hour"]+(HK_valid["Day"]-1)*24

plt.rcParams['axes.facecolor'] = 'gray'
plt.plot(HK_valid["2018_12_Time"],HK_valid["Level"],color='orange',label="Sea Level in December, 2018, HK")
plt.xlabel("Time: unit-Hour")
plt.ylabel("Sea Level")
plt.title("Sea Level Variation in December, 2018, HK")
plt.legend()
plt.show()

fs = 1/(60*60)
freq, power = signal.periodogram(HK_valid["Level"],fs)
plt.semilogy(freq, power)
plt.xlim(0, 0.000006)
plt.ylim(1e-2,1e11)
plt.xlabel("Frequency: unit-Hz")
plt.ylabel("Power")
plt.title("Sea Level Variation in December, 2018, HK")
plt.legend()
plt.show()

```
