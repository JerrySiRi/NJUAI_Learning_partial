### 1. Python Lists to 2-D NumPy Arrays

```python
# Import numpy
import numpy as np

# define a list named baseball
baseball = [[180, 78.4],
            [215, 102.7],
            [210, 98.5],
            [188, 75.2]]

# Create a 2D Numpy array from the Python list baseball, name it "np_baseball"

np_baseball = np.array(baseball)

# print out the dimension of np_baseball

print(np_baseball.ndim)

# Print out the type of np_baseball

print(type(np_baseball))

# Print out the shape of np_baseball

print(np_baseball.shape)

# Print out the second row of np_baseball

print(np_baseball[1,:])

# print out the first colume of baseball

print(np_baseball[:,0])

# what is the value of np_baseball[2,1]?

print(np_baseball[1,0])

```

### 2. Creating 2-D NumPy arrays

* Now, Each row represnts one earthquake
* The column data goes as "Latitude", "Longitude", "Depth", "Magnitude"
* Make a histgram to show the distributions of the magnitudes of all the earthquakes
* Plot the relationship between "Depth" and "Latitude", do you find any correlations?
* Plot the relationship between "Depth" and "Magnitude", What do you find?

```python
# print(help(np.loadtxt))
import matplotlib.pyplot as plt
EQ = np.loadtxt('earthquake.csv', delimiter=',', skiprows=1)
latitude = EQ[:,0]
longtitude = EQ[:,1]
depth = EQ[:,2]
magnitude = EQ[:,3]

plt.hist(longtitude,bins=50)
plt.show()

print('depth vs. latitude')
plt.subplot(1,2,1)
plt.title('depth vs. latitude --- scatter')
plt.scatter(latitude,depth)
plt.show()
plt.subplot(1,2,2)
plt.title('depth vs. latitude --- plot')
plt.plot(latitude,depth)
plt.show()

print('depth vs. magnitude')
plt.subplot(1,2,1)
plt.title('depth vs. magnitude --- scatter')
plt.scatter(magnitude,depth)
plt.show()
plt.subplot(1,2,2)
plt.title('depth vs. magnitude --- plot')
plt.plot(magnitude,depth)
plt.show()

```
